```python
from sklearn.feature_extraction.text import CountVectorizer
import pickle

corpus = []

with open("parsed_review_data.pkl", "rb") as f:
    try:
        while True:
            value = pickle.load(f)
            li = []
            for i in value:
                tmp = i.split("/")
                li.append(tmp[0])
            
            str = ' '.join(li)
            print(str)
            corpus.append(str)
            
            
            

    except EOFError:
        print("파일 읽기 종료")
```
